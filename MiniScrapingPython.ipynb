```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Scoville_scale"
page = requests.get(url) #getting information from this url
soup = BeautifulSoup(page.text, 'html.parser') #pulling info from the url, html is parser
```

```python
print(soup) #pulling from page
```

```python
#finding our table
soup.find('table')
```

```python
soup.find_all('table')[1] #getting table
```

```python
soup.find('table',class_='wikitable') #specify using find class
```

```python
table = soup.find_all('table')[1] #this will be our table we're working with
```

```python
#print(table) #printing table
```

```python
spice_titles=table.find_all('th') #this is where the titles of table is
```

```python
spice_table_titles=[title.text.strip() for title in spice_titles]
print(spice_table_titles) #printing the 3 table titles --> Scoville heat units, Chemical, and "Ref"
```

```python
import pandas as pd
df=pd.DataFrame(columns = spice_table_titles) #put the titles into a dataframe
df
```

```python
col_data = table.find_all('tr') #getting actual row data
```

```python
for row in col_data[1:]:
  row_data= row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data] #stripping to help w readability, looping through the code

  #checking length of df each time looping, and then putting in data in next position
  #doing this because otherwise it will only save each 1 row at a time instead of appending
  length=len(df)
  df.loc[length] = individual_row_data
```

```python
df #pulling new dataframe with data and titles
```

```python
df.to_csv('Spices_Chemicals.csv',index=False) #exporting dataframe to a csv file,
# index=False --> won't include the indexes into csv (0-9)
```
